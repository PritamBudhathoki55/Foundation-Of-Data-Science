## Eigen Values and Eigen Vectors:
Eigenvalues and Eigenvectors are fundamental concepts in linear algebra and play a major role in data science, especially in dimensionality reduction, feature extraction, and data transformation. It is important for PCA(Principal Component Analysis) where data varies the most. Eigenvectors show where the data is pointing,
Eigenvalues show how important that direction is. It is also used in preprocessing before ML model.



Q.Compute the eigenvalues and eigenvectors of the matrix:
C=[[4    1], [​−2       1]​]


```python
C = np.array([[4, 1],
              [-2, 1]])

eigenvalues, eigenvectors = np.linalg.eig(C)

print("Matrix A:")
print(C)
print("\nEigenvalue (first):")
print(eigenvalues[0])
print("\nCorresponding Eigenvector:")
print(eigenvectors[:, 0])
```

## Car Price Prediction using Linear Regression

You are given a dataset of cars with the following features:Number of cylinders (x1​), Engine size in liters (x2​), Horsepower (x3)

car 1:

x1: 4

x2:  1.5

x3: 100

and price: 20


car 2:

x1: 6

x2:  2.0

x3: 150

and price: 30

Then, find the price of a car 3 with features

x1: 8

x2:  3.0

x3: 200

price:?



```python
# Feature matrix (2 cars, 3 features)
# x1 = cylinders, x2 = engine size, x3 = horsepower
X = np.array([
    [4, 1.5, 100],   # Car 1
    [6, 2.0, 150]    # Car 2
])

# Price vector
y = np.array([20, 30])
XXT=X @ X.T
print("\nX X^T:\n",XXT)

XXT_invert=LA.inv(XXT)
print("\n(X*X^T)^-1\n",XXT_invert)

beta=X.T @ XXT_invert @ y
print("\nRegression Coefficient(beta):\n",beta)

car3_features = np.array([8, 3.0, 200])
car3_price = car3_features @ beta
print(f"\nPredicted price for Car 3: {car3_price:.2f}")
```

<p>In this topic, We constructed feature Matrix "X" and Price Vector "y".I used pesudo-inverse because this gives the minimum-norm solution:β vector with smallest magnitude that fits the data.</p>
The pseudo-inverse (also called Moore-Penrose inverse) is a generalization of the matrix inverse for non-square or singular matrices Computed regression coefficient (Beta) and predicted the price of new car usin dot product.

<b>Result: 

Predicted Price of Car3: </b>=40
