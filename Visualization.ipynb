## Line Chart

```python

# given age range (27â€“59)
Age = [27, 32, 37, 42, 47, 52, 57]
Stress_Level = [3, 4, 5, 4, 6, 5, 7]

# create the line chart
plt.plot(Age, Stress_Level)

# add title and labels
plt.title("Stress Level by Age")
plt.xlabel("Age")
plt.ylabel("Stress Level")

plt.show()

```

In this Line plot,It shows that the stress usually increases with age. From the chart, It is clearly visible that younger adults (27-37) experiences lower stress whereas middle aged peoples(42-57) face higher and more fluctuating stress levels. It shows that life responsibilities and certain events in life cause variations in stress over the time period.

Therefore,Age is positively correlated with stress level in general, but the relationship between them in not perfectly linear. Life events and responisibilites at certain age cause ups and downs in stress levels of human beings.



## Bar Plot

```python
bmi_data = ['Underweight', 'Normal', 'Overweight', 'Obese',
            'Normal', 'Overweight', 'Normal', 'Obese', 'Normal']

# Load data into DataFrame
load = pd.DataFrame(bmi_data, columns=['BMI Category'])

# Bar Plot
load['BMI Category'].value_counts().plot(kind='bar')

plt.title('Bar Plot of BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Number of People')
plt.show()
```

This bar plot represents the distribution of participants across different BMI categories. The X-axis shows the BMI categories: Underweight, Normal, Overweight, and Obese, while the Y-axis shows the number of participants in each category.

From the plot, we can observed that the Normal BMI group has the highest number of participants, whereas the Underweight group has the lowest. This indicates that most participants fall within a healthy BMI range.
This bar plot helps in viusally comparing categorical data,making it easier to identify which BMI category is more relevant among the number of participants.

## Box Plot

```python
sns.boxplot(y=load['Sleep_Duration'], color='skyblue')
plt.title('Box Plot of Sleep Duration')
plt.ylabel('Hours of Sleep')
plt.show()
```

This box plot shows distribution of sleep duration of participants.As we know, The box represents the interquartile range i.e. 50% of its value, line inside the box shoes the median and the whisker shows minimum and maximum values. Dot outside and inside the whisker shows the outliers.
Therefore, It helps to find the various slleep duration of the participants.

## Pie Chart

```python
load['Gender'].value_counts().plot(kind='pie',autopct='%0.2f%%')
plt.title('Gender Distribution')
plt.show()
```

This pie chart represents the gender distribution for the experiment. It shows that there is almost equal number of gender distribution.Just by margin,There are little more male for this experiment.
In conclusion,It can be said that both males and females are nearly equally represented in the dataset which makes it suitable for comparative analysis without any biasness for experiment related to sleep health and lifestyle factors across genders.

## Histogram

```python
plt.hist(load['Sleep_Duration'])
plt.title('Box Plot of Sleep_Duration among people')

```

This is a visual representation of Histogram, It shows duration of sleep among the people.It is clearly seen that most people sleep between 6 to 8 hours daily. Very few people sleeping less than 6 hours or more than 8.5 hours.

```python
sns.displot(load['Sleep_Duration'],kind="hist",kde=True) 
plt.title('Histogram of Sleep Duration using KDE')
plt.xlabel('Hours of Sleep')
plt.ylabel('No. of People')
plt.show()
```

This is also histogram chart but we used kde also called Kernel Density Estimation which is used to provide a smooth estimate of data distribution.
It helps understanding the shape of given dataset. It helps to find the maximun and minimun value in a dataset easily.
In this chart,KDE curve peaks around 7-7.5 hours and most individual sleeping between 6 to 8 hours.

## Scatter Plot
